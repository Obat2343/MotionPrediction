```python
import numpy as np
import cv2

size = 15
vel_map = np.zeros((2,size,size))

for i in range(size):
    for j in range(size):
        vel_map[1,i,j] = i - ((size - 1)/2)
        vel_map[0,i,j] = j - ((size - 1)/2)

mag, ang = cv2.cartToPolar(vel_map[0], vel_map[1])
hsv = np.zeros((size,size,3))
hsv[...,1] = 255
hsv[:,:,0] = ang*180/np.pi/2
hsv[:,:,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)

rgb = cv2.cvtColor(hsv.astype('uint8'),cv2.COLOR_HSV2RGB)
```

```python
from PIL import Image
rgb_image = Image.fromarray(rgb)
rgb_image.resize((256,256))
```
